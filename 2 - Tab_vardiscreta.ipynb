1. Importar as bibliotecas Pandas (pd) e math

```python
import pandas as pd
import numpy as np
import math
```

2. Definir a variável "arquivo" pelo nome da planilha.


```python
arquivo='dados_projeto1.xlsx'
```

3. Ler a planilha no Python. Escolha um nome para armazenar. Escolheremos df

```python
df=pd.read_excel(arquivo)
df
```

4.Agrupando os valores quantitativos a partir da variável selecionada.Escolhemos Idade (nome da coluna no arquivo df)

```python
df1=df.Idade.value_counts()
df1
```

5. Ordenando os valores da variável

```python
df1=df.Idade.value_counts().sort_index()
df1
```

6. Contando o total de elementos

```python
totaldf1=df1.sum()
totaldf1
```

7. Montando a tabela de frequências da variável quantitativa discreta

7.1. Criar um DataFrame para a coluna desejada (dadosvquanti no exemplo)









```python
dadosvquanti = pd.DataFrame(df1).reset_index()
dadosvquanti.columns = ['Idade', 'Frequência']
```

```python
dadosvquanti
```

7.2 Calcular a frequência relativa em % com 2 casas decimais

```python
dadosvquanti['Frequência Relativa (%)'] = round(dadosvquanti['Frequência'] * 100 / totaldf1, 2)
```

7.3 Criando uma nova tabela, renomeando as colunas

```python
tabela=pd.DataFrame({"Idade":dadosvquanti['Idade'],"Frequência":dadosvquanti['Frequência'],"Frequência Relativa (%)":dadosvquanti['Frequência Relativa (%)']})
```

```python
tabela
```

**Montando as colunas das frequências acumuladas**

7.4 Frequências acumuladas

```python
dadosvquanti['Frequência Acumulada']=np.cumsum(dadosvquanti['Frequência'])
dadosvquanti
```

```python
dadosvquanti['Frequência Relativa Acumulada (%)']=np.cumsum(dadosvquanti['Frequência Relativa (%)'])
dadosvquanti
```

Ajustando a Tabela com colunas das frequências acumuladas

```python
tabela=pd.DataFrame({"Idade":dadosvquanti['Idade'],"Frequência":dadosvquanti['Frequência'],"Frequência Relativa (%)":dadosvquanti['Frequência Relativa (%)'], "Frequência Acumulada":dadosvquanti['Frequência Acumulada'],"Frequência Relativa Acumulada (%)":dadosvquanti['Frequência Relativa Acumulada (%)']})
```

```python
tabela
```
